## <font color='264CC7'> Funciones auxiliares </font>

Usaremos algunas funciones para registrar iteraciones y calcular errores aproximados.

```python
def error_aproximado(x_actual, x_anterior):
    if x_anterior is None:
        return None
    return abs(x_actual - x_anterior)

def error_aproximado_relativo(x_actual, x_anterior):
    if x_anterior is None or x_actual == 0:
        return None
    return abs(x_actual - x_anterior) / abs(x_actual)

def hay_cambio_de_signo(funcion, a, b):
    return funcion(a) * funcion(b) < 0
```

## <font color='264CC7'> 3. Método de punto fijo </font>

El método de punto fijo transforma una ecuación `f(x)=0` en una ecuación equivalente:

$$
x=g(x).
$$

Luego genera aproximaciones con:

$$
x_{i+1}=g(x_i).
$$

</div>

```python
def punto_fijo(g, x0, tolerancia=1e-6, max_iter=100):
    filas = []

    for i in range(1, max_iter + 1):
        # TODO: calcula la siguiente aproximación.
        x1 = None
        err = error_aproximado(x1, x0) if x1 is not None else None

        filas.append({
            'iteracion': i,
            'x_anterior': x0,
            'x': x1,
            'error_aprox': err,
        })

        # TODO: criterio de parada.

        x0 = x1

    return pd.DataFrame(filas)

# Ejemplo: x^3 + x - 1 = 0, con x = 1 - x^3.
g1 = lambda x: 1 - x**3
# tabla_pf = punto_fijo(g1, x0=0.5, tolerancia=1e-5, max_iter=50)
# tabla_pf.tail()
```

## <font color='264CC7'> 4. Método de Newton-Raphson </font>

Newton-Raphson usa la recta tangente a la función. Su fórmula es:

$$
x_{i+1}=x_i-\frac{f(x_i)}{f'(x_i)}.
$$

<div style="background-color: #edf1f8; border-color: #264CC7; border-left: 5px solid #264CC7; padding: 0.5em;">
<strong>Ejercicio 7:</strong><br>
Completa el método de Newton-Raphson y aplícalo a <code>f(x)=x^2-2</code> con <code>x0=1</code>.
</div>

```python
def newton_raphson(funcion, derivada, x0, tolerancia=1e-6, max_iter=100):
    filas = []

    for i in range(1, max_iter + 1):
        fx = funcion(x0)
        dfx = derivada(x0)

        if dfx == 0:
            raise ZeroDivisionError('La derivada se anuló durante la iteración.')

        # TODO: calcula x1 con la fórmula de Newton-Raphson.
        x1 = None
        err = error_aproximado(x1, x0) if x1 is not None else None

        filas.append({
            'iteracion': i,
            'x_anterior': x0,
            'x': x1,
            'f(x_anterior)': fx,
            "f'(x_anterior)": dfx,
            'error_aprox': err,
        })

        # TODO: criterio de parada.

        x0 = x1

    return pd.DataFrame(filas)

h = lambda x: x**2 - 2
dh = lambda x: 2*x

# tabla_newton = newton_raphson(h, dh, x0=1, tolerancia=1e-8, max_iter=20)
# tabla_newton
```

## <font color='264CC7'> 5. Método de la secante </font>

La secante aproxima la derivada usando dos puntos anteriores:

$$
x_{i+1}=x_i-\frac{f(x_i)(x_i-x_{i-1})}{f(x_i)-f(x_{i-1})}.
$$

<div style="background-color: #edf1f8; border-color: #264CC7; border-left: 5px solid #264CC7; padding: 0.5em;">
<strong>Ejercicio 9:</strong><br>
Completa el método de la secante. Aplícalo a <code>f(x)=x^2-2</code> con <code>x0=1</code> y <code>x1=2</code>.
</div>

```python
def secante(funcion, x0, x1, tolerancia=1e-6, max_iter=100):
    filas = []

    for i in range(1, max_iter + 1):
        fx0 = funcion(x0)
        fx1 = funcion(x1)
        denominador = fx1 - fx0

        if denominador == 0:
            raise ZeroDivisionError('El denominador de la fórmula de secante se anuló.')

        # TODO: calcula x2 con la fórmula de la secante.
        x2 = None
        err = error_aproximado(x2, x1) if x2 is not None else None

        filas.append({
            'iteracion': i,
            'x_anterior': x0,
            'x_actual': x1,
            'x': x2,
            'f(x)': funcion(x2) if x2 is not None else None,
            'error_aprox': err,
        })

        # TODO: criterio de parada.

        x0 = x1
        x1 = x2

    return pd.DataFrame(filas)

# tabla_secante = secante(h, 1, 2, tolerancia=1e-8, max_iter=20)
# tabla_secante
```

## <font color='264CC7'> 6. Comparación de métodos </font>

Completa la siguiente tabla conceptual antes de comparar resultados numéricos.

| Método | Datos iniciales | ¿Requiere cambio de signo? | ¿Requiere derivada? | Velocidad esperada |
|---|---|:---:|:---:|---|
| Bisección | Intervalo `[a,b]` | Sí | No | Lineal, razón aproximada `1/2` |
| Falsa posición | Intervalo `[a,b]` | Sí | No | Generalmente lineal |
| Punto fijo | `x0` y función `g` | No | No | Lineal, razón `|g'(r)|` |
| Newton-Raphson | `x0` y `f'` | No | Sí | Cuadrática, orden `2` |
| Secante | `x0`, `x1` | No | No | Superlineal, orden `1.618` |
